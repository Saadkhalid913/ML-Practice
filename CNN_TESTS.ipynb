```python
import torch
import torch.nn as nn
import torchvision.datasets as datasets
import torchvision.transforms as transforms 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
mean_grey = 0.1207
grey_std = 0.3081

# refer back 
transformations =transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=mean_grey, std=grey_std)])

Train_dataset = datasets.MNIST("/data", train=True, transform=transformations, download=True)
Test_dataset = datasets.MNIST("/data", train=False, transform=transformations, download=False)
```

```python
train_loader = torch.utils.data.DataLoader(Train_dataset, batch_size=100, shuffle=True)
test_loader = torch.utils.data.DataLoader(Test_dataset, batch_size=100, shuffle=False)
```

```python
class CNN(nn.Module):
  def __init__(self):
    super(CNN, self).__init__()
    self.cnn1 = nn.Conv2d(1, 8, kernel_size=3, stride=1, padding=1)
    self.batchnorm1 = nn.BatchNorm2d(8)
    self.relu = nn.ReLU()
    self.maxpool1 = nn.MaxPool2d(kernel_size=2)
    self.cnn2 = nn.Conv2d(in_channels=8, out_channels=32, kernel_size=5, stride=1, padding=2)
    self.batchnorm2 = nn.BatchNorm2d(32)
    self.maxpool2 = nn.MaxPool2d(kernel_size=2) # change 

    self.fc1 = nn.Linear(7 * 7 * 32, 600) # change 
    self.dropout = nn.Dropout(p = 0.5)
    self.fc2 = nn.Linear(600, 10) # change 

  def forward(self, x):
    out = self.cnn1(x)
    out = self.batchnorm1(out)
    out = self.relu(out)
    out = self.maxpool1(out)

    out = self.cnn2(out)
    out = self.batchnorm2(out)
    out = self.maxpool2(out)

    out = out.view(-1, 1568)

    out = self.fc1(out)
    out = self.relu(out)
    out = self.dropout(out)
    out = self.fc2(out)

    return out 


n = CNN()

loss_func = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(n.parameters(), lr = 0.01)


```

```python
train_loss = []
test_loss = []
train_accuracy = []
test_accuracy = []

for i in range(1):
  correct = 0
  iterations = 0
  iter_loss = 0.0 
  n.train()

  for j , (features, labels) in enumerate(train_loader):
    out = n.forward(features)
    loss = loss_func(out, labels)
    iter_loss += loss
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    _, predictions = torch.max(out, dim=1)
    print(j + 1, "Correct: ", (labels == predictions).sum().item() , "%")
  
    correct += (labels == predictions).sum().item()

  accuracy = (correct / (100 * len(train_loader))) * 100
  print(f'Accuracy {accuracy}%')



```
